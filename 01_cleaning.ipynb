```python
import pandas as pd
import numpy as np

def clean_data(df):

    df = df.copy()

    # Drop useless columns
    df.drop(['Ticket', 'Cabin', 'Name'], axis=1, inplace=True)

    # Fill missing
    df['Age'].fillna(df['Age'].median(), inplace=True)
    df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

    # Remove duplicates
    df = df.drop_duplicates()

    # Fix data types
    df['Sex'] = df['Sex'].astype('category')
    df['Pclass'] = df['Pclass'].astype('category')

    # Handle outliers in Fare
    upper = df['Fare'].quantile(0.99)
    df['Fare'] = np.where(df['Fare'] > upper, upper, df['Fare'])

    return df


df = pd.read_csv("/content/train.csv")
df_clean = clean_data(df)

# Checks
assert df_clean.isnull().sum().sum() == 0
assert (df_clean['Fare'] >= 0).all()
assert df_clean.shape[0] > 0

df_clean.to_csv("/content/cleaned.csv", index=False)
```
